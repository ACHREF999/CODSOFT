```python
data = pd.read_csv('/kaggle/input/bank-customer-churn-prediction/Churn_Modelling.csv')
data = data.drop(['RowNumber','CustomerId','Surname'],axis=1)
# What we are going to do is to encode the data we have , geography,gender that is and change bools if any to 0/1
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
label_encoder = LabelEncoder()
min_max_scaler = MinMaxScaler()


data['Gender'] = label_encoder.fit_transform(data['Gender'])
data['Geography'] = label_encoder.fit_transform(data['Geography'])

scaled_data = data
scaled_data[['Age','Balance','EstimatedSalary','CreditScore','Tenure']] = min_max_scaler.fit_transform(data[['Age','Balance','EstimatedSalary','CreditScore','Tenure']])
scaled_data = pd.DataFrame(scaled_data, columns=data.columns)
scaled_data

X= scaled_data.loc[:,~scaled_data.columns.isin(['Exited'])]
Y = data['Exited']
X['Tenure'].unique()
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)


svm_model = SVC(kernel='linear')  
svm_model.fit(X_train, y_train)

y_pred = svm_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
cs_report = classification_report(y_test, y_pred)

print("Accuracy:", accuracy)
print("Classification Report:\n", cs_report)
```

```python
svm_model = SVC(kernel='rbf')  
svm_model.fit(X_train, y_train)

y_pred = svm_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
cs_report = classification_report(y_test, y_pred)

print("Accuracy:", accuracy)
print("Classification Report:\n", cs_report)
```
